```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer


```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df['Type'])

vectorize = TfidfVectorizer(stop_words='english')
x = vectorize.fit_transform(df['Description'])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x.shape, len(label_encoder.classes_)
```

```python
#Naive Bayes
nb_model = MultinomialNB()
nb_model.fit(x_train, y_train)
nb_pred = nb_model.predict(x_test)
```

```python
#SVM
svm_model = LinearSVC()
svm_model.fit(x_train, y_train)
svm_preds = svm_model.predict(x_test)
```

```python
nb_report = classification_report(y_test, nb_pred, target_names=label_encoder.classes_)
svm_report = classification_report(y_test, svm_preds, target_names=label_encoder.classes_)
print("Naive Bayes Report:\n", nb_report)
print("SVM Report:\n", svm_report)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt



# Accuracy scores
nb_accuracy = accuracy_score(y_test, nb_pred)
svm_accuracy = accuracy_score(y_test, svm_preds)

print(f"Naive Bayes Accuracy: {nb_accuracy:.2f}")
print(f"SVM Accuracy: {svm_accuracy:.2f}")

# Confusion Matrix - Naive Bayes
cm_nb = confusion_matrix(y_test, nb_pred)
disp_nb = ConfusionMatrixDisplay(confusion_matrix=cm_nb, display_labels=label_encoder.classes_)
disp_nb.plot(cmap="Blues")
plt.title("Confusion Matrix - Naive Bayes")
plt.show()

# Confusion Matrix - SVM
cm_svm = confusion_matrix(y_test, svm_preds)
disp_svm = ConfusionMatrixDisplay(confusion_matrix=cm_svm, display_labels=label_encoder.classes_)
disp_svm.plot(cmap="Greens")
plt.title("Confusion Matrix - SVM")
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt



# Create Priority column based on Benefit/Cost ratio
df['Priority'] = pd.cut(df['Benefit'] / df['Cost'], bins = [0, 1.5, 3, float('inf')], labels=['Low', 'Medium', 'High'], duplicates='drop')

# Select features
features = df[['Region', 'Complexity', 'Status','Completion%', 'Cost', 'Benefit']]
features = pd.get_dummies(features)  # One-hot encode categorical variables

# Encode the target
le = LabelEncoder()
target = le.fit_transform(df['Priority'])

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf.predict(X_test)
print("ðŸ”Ž Random Forest Classification Report:")
#print(classification_report(y_test, y_pred, target_names=le.classes_))
print("ðŸŽ¯ Accuracy:", accuracy_score(y_test, y_pred))

# Feature importance
importances = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
print("\nTop 10 Features:")
print(importances.head(10))

# Plot
plt.figure(figsize=(10, 6))
importances.head().plot(kind='barh', color='green')
plt.title("Top 10 Feature Importances (Random Forest)")
plt.xlabel("Importance Score")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

workload balancing ML Approach

```python
# Filter in-progress projects (workload)
in_progress = df[df['Status'] == 'in - Progress']
#in_progress = df[df['Status'] == 'in - progress']
#Aggregate current workload by manager
manager_workload = in_progress.groupby('Manager').agg(
    total_cost=('Cost', 'sum'),
    total_benefit = ('Benefit', 'sum'),
    project_count=('Cost', 'count')

).reset_index()
#Prepare data for ML model
features = df[['Region', 'Department', 'Complexity', 'Phase', 'Manager', 'Completion%', 'Benefit']]
features = pd.get_dummies(features)
target = df['Cost']
print(manager_workload)
```

```python
# Normalize status values to ensure consistent matching
df['Status'] = df['Status'].str.strip().str.lower()

# Filter completed projects (after normalization)
completed = df[df['Status'] == 'completed']

# Aggregate workload summary by manager
manager_workload = completed.groupby('Manager').agg(
    total_cost=('Cost', 'sum'),
    total_benefit=('Benefit', 'sum'),
    project_count=('Cost', 'count')
).reset_index()

print(manager_workload)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
#Train Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
#Predict future workload of all managers
future_projects = df[df['Status'] != 'in - progress']
future_features = pd.get_dummies(future_projects[['Region', 'Department', 'Complexity', 'Phase', 'Manager', 'Completion%', 'Benefit']])
# Align columns
future_features = future_features.reindex(columns=X_train.columns, fill_value=0)

predicted_costs = rf_model.predict(future_features)
future_projects = future_projects.copy()
future_projects['Predicted_Cost'] = predicted_costs

#Recommend manager with lowest projected load
recommendation = future_projects.groupby('Manager')['Predicted_Cost'].sum().reset_index()
recommendation = recommendation.merge(manager_workload, on='Manager', how='left').fillna({'total_cost': 0})
recommendation['Projected_Total_Load'] = recommendation['total_cost'] + recommendation['Predicted_Cost']
least_loaded_manager = recommendation.loc[recommendation['Projected_Total_Load'].idxmin()]

print("ðŸ“Œ Recommended Manager for Next Assignment:")
print(least_loaded_manager)
```

 GridSearchCV for hyperparameter tuning.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Define the parameter grid for Random Forest
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Create a RandomForestRegressor
rf = RandomForestRegressor(random_state=42)
```

```python
grid_search = GridSearchCV (
    estimator = rf,
    param_grid = param_grid,
    cv = 3,
    n_jobs = -1,
    verbose = 2
)
```

```python
grid_search.fit(X_train, y_train)
print("âœ… Best Parameters Found:")
print(grid_search.best_params_)
best_rf = grid_search.best_estimator_
```

```python
y_pred = best_rf.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(mse)
rmse  = np.sqrt(mse)
print(f"Mean_squared_error: {rmse : .2f}")
```

```python
import pandas as pd
import dash
from dash import dcc, html
import plotly.express as px
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# ML Model Setup
features = df[['Region', 'Department', 'Complexity', 'Phase', 'Manager', 'Completion%', 'Benefit']]
features = pd.get_dummies(features)
target = df['Cost']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

model = RandomForestRegressor(random_state=42)
param_grid = {
    'n_estimators': [100],
    'max_depth': [20],
    'min_samples_split': [2],
    'min_samples_leaf': [1]
}
grid_search = GridSearchCV(model, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


# Dash App
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("ðŸ“Š Project Summary Dashboard"),

    html.H2("Manager Workload Summary"),
    dcc.Graph(
        figure=px.bar(manager_workload, x='Manager', y='total_cost', title='Total Cost by Manager')
    ),
    dcc.Graph(
        figure=px.bar(manager_workload, x='Manager', y='total_benefit', title='Total Benefit by Manager')
    ),

    html.H2("ML Model Performance"),
    html.Ul([
        html.Li(f"Best Parameters: {grid_search.best_params_}"),
        html.Li(f"Test MSE: {mse:,.2f}"),
        html.Li(f"Test RÂ² Score: {r2:.2f}")
    ]),


])

if __name__ == '__main__':
    app.run(debug=True)
```
